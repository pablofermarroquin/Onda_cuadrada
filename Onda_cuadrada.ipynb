```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.patches as mpatches
```

```python
#Amortiguado forzado con onda cuadrada

b=1
k = 10
m=8
beta = b/(2*m)
wo2=k/m

t= np.linspace(0, 35, 300, endpoint = True) 
#xc=0.47*np.exp(-0.16*t)*np.cos(8.67*t-6.12)
xc=np.exp(-beta*t)*(np.cos(((-beta**2+wo2)**0.5)*t))

#===============================================
#crear serie de fourier
terminos=100
bn=[]
xp=t-t
for n in range (1,terminos+1,1):
  bi=-(2/(np.pi*n))*(np.cos(n*np.pi)-1)
  bn.append(bi)
  xp+=bi*np.sin(n*t)
xp+=1
xTOT=xc+xp

plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.plot(t,xc,c="g")
plt.plot(t,xp,"b")
plt.plot(t,xTOT, "r")
plt.title("\nPosición vrs. tiempo") 

#Leyenda

id_xc= mpatches.Patch(color='green', label='Xc(t)')
id_xp= mpatches.Patch(color='blue', label='Xp(t)')
id_xTOT= mpatches.Patch(color='red', label='Xc(t)+ Xp(t)')
plt.legend(handles=[id_xc, id_xp,id_xTOT])

plt.show
```
